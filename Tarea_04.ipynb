### Problema 1:
#### Determina qué ocurre cuando creamos un objeto Series a partir de un diccionario, pero adicionalmente proveemos el argumento opcional index. ¿De qué forma podemos utilizar esta propiedad?

```python
# Creación de diccionario
diccionario = {'a': 64,'b': 23,'c': 13,'d': 85,'e': 14}

# Creación de series con orden de índice específico
serie = pd.Series(diccionario, index=['c', 'd', 'b', 'e', 'a'])
print(serie)
```

Los índices **c, d, b, e y a** se ordenan segun el argumento index del método Series de pandas. Con el ejemplo anterior podemos entender como el argumento index sirve para especificar el orden particular del objeto Serie.

### Problema 3: 
#### Investiga para qué pueden ser utilizados los arreglos Structured Arrays de NumPy.

Los Structured Arrays de NumPy se utilizan para agrupar datos de diferentes tipos y tamaños. La matriz de estructura utiliza “contenedores” de datos llamados campos; cada campo puede contener datos de cualquier tipo y tamaño. Se puede acceder a los elementos de la matriz con la ayuda de la notación de puntos.

Para poder hacer uso de estos arreglos se debe considerar los siguientes requisitos:

•	Todas las estructuras en la matriz deben tener el mismo número de campos.

•	Todas las estructuras deben tener los mismos nombres de campo.

```python
import numpy as np
```

```python
alumnas = np.array([('Daniela', 24, 72.1), ('Melissa', 25, 69.2)],
       dtype=[('nombre', (np.str_, 10)), ('edad', np.int32), ('peso', np.float64)])
              
# Sortear por nombre
b = np.sort(alumnas, order='nombre')
print('Sortear conforme a la inicial del nombre:', b)
 
# Sortear por edad
b = np.sort(alumnas, order='edad')
print('\nSortear conforme a la edad:', b)
```

### Problema 4:
#### Investiga las operaciones **isnull, notnull, dropna y fillna** de Pandas, así como el valor **pd.NA**.

#### isnull()
El argumento **isnull()** devuelve un objeto DataFrame donde todos los valores se reemplazan con un valor booleano True para valores NULL y, de lo contrario, False.

```python
# Creación de data frame desde un archivo .csv
data = pd.read_csv("employees.csv")
 
# Creando series booleanas True para valores NaN
bool_series = pd.isnull(data["Team"])

data[bool_series]
```

Al hacer uso del argumento isnull(), solo se muestran las filas que tienen Team=NULL.

#### notnull()
La función **notnull()** devuelve valores booleanos para reflejar si los valores evaluados son nulos (false) o no nulos (true); es un alias del método pandas .notna.

```python
# Creando series booleanas False para valores NaN
bool_series = pd.notnull(data["Gender"])
 
# Los datos mostrados son exclusivos de Gender sin NaN
data[bool_series]
```

Al hacer uso del argumento notnull(), solo se muestran las filas cuyo Gender no es = NULL.

#### dropna()
La función **dropna()** se usa para descartar/eliminar valores faltantes de filas y columnas en un data frame.

```python
sinna = data.dropna()
sinna
```

Al hacer uso del argumento dropna(), solo se muestran las filas que no tienen ningún valor faltante/NULL).

#### fillna()
El método **fillna** permite sustituir los valores nulos de una estructura pandas por otro valor según ciertos criterios: pueden sustituirse por un valor concreto o bien puede utilizarse el anterior o posterior valor no nulo.

```python
fillna = data.fillna("aquí no hay datos faltantes")
fillna
```

Al hacer uso del argumento fillna(), todo los valores NA se reemplazan por "aquí no hay datos faltantes", como se definió arriba.

#### pd.NA
Un valor **pd.NA** es experimental y está disponible para representar valores faltantes escalares. De utiliza mayormente en los tipos de datos de cadena dedicados, booleanos y enteros anulables como indicador de valor faltante.

El objetivo de pd.NA es proporcionar un indicador "ausente" que se pueda usar de manera consistente en todos los tipos de datos (en lugar de np.nan, Ninguno o pd.NaT que se basan según el tipo de datos).
